c) Eine ganz andere Art, eine Funktion in zwei Variablen darzustellen, besteht darin, die eine Variable quasi als Zeit zu interpretieren. Erstellen Sie eine Animation der gleichen Funktion wie im Aufgabenteil a), indem Sie x als Ortsvariable und y als Zeitvariable auffassen. Sie bekommen dabei eine 2Dâ€“ und nicht mehr eine 3Dâ€“Darstellung (Hinweis: FuncAnimation von dem matplotlib.animation). Stellen Sie die Animation auch fÃ¼r die Intervalle x = âˆ’15..15 und y = âˆ’15..15 dar.




```python
%matplotlib notebook

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], '-')
xdata = np.arange(-15, 15, 0.2)

def F(y):
    liste = []
    for i in xdata:
        if i == 0 and y == 0:
            liste.append([1])
        else:
            liste.append([sin(sqrt(i**2 + y))/sqrt(i + y**2)])
    return liste

def init():
    ax.set_xlim(-15, 15)
    ax.set_ylim(-1, 1)
    ax.grid()
    return ln,

def update(frame):

    ydata = F(frame)
    ln.set_data(xdata, ydata)
    return ln,

step = 2
frames = np.arange(-15, 15, step)

ani = FuncAnimation(fig, update, frames,init_func=init, interval=10)

plt.show()
```
